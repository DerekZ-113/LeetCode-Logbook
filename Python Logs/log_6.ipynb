```python
# Players and Hats

# Parameters:
# total_players 1 to 8
# total_hats 1 to 6
# owned_hat_length 1 to 6, every player has the same number of owned hats
# hat_list: 2D list index is the player, values are the hats owned by that player
# points of each that range 1 to 100

# find all the possible distributions of hats to players when each player has one distinct hat
# calculate the max score from all distributions, it's the multiply from all hats in one path

# Example:
# total_players = 3
# total_hats = 3
# owned_hat_length = 2
# hat_list = [[0,1],[1,2],[0,2]]
# hat_points = [1,2,3]
# All possible distributions:

# Max score = 60   
# Output: [3, 60] first is the number of distributions, second is the max score
```

```python
def calculate(total_players, total_hats, owned_hat_length, hat_list, hat_points):
    if total_players > total_hats:
        return [0, 0]

    exisiting_hats = set()
    def dfs(player, current_ways, current_score):
        if player == total_players:
            return [current_ways, current_score]

        for hat in hat_list[player]:
            if hat not in exisiting_hats:
                exisiting_hats.add(hat)
                if len(exisiting_hats) == total_players:
                    current_ways += 1
                    # calculate the score
                    max_score = current_score
                    current_score = 1
                    for wore_hat in exisiting_hats:
                        current_score *= hat_points[wore_hat]
                    current_score = max(current_score, max_score)
                    [current_ways, current_score] = dfs(player + 1, current_ways, current_score)
                    existin_hats.remove(hat)

        return [current_ways, current_score]

    result = [0, 0]
    result = dfs(0, 0, 1)
    return result
```

```python
def calculate(total_players, total_hats, owned_hat_length, hat_list, hat_points):
    if total_players > total_hats:
        return [0, 0]

    existing_hats = set()  # Fixed typo: exisiting -> existing
    total_ways = 0
    
    def dfs(player, current_ways, current_score):
        if player == total_players:
            # Calculate final score when all players have hats
            score = 1
            for wore_hat in existing_hats:
                score *= hat_points[wore_hat]
            return [current_ways + 1, max(score, current_score)]  # Count this way and update max score

        # Track the best results from all branches
        best_ways = current_ways
        best_score = current_score
        
        for hat in hat_list[player]:
            if hat not in existing_hats:
                existing_hats.add(hat)
                # Recursive call and collect results
                [ways, score] = dfs(player + 1, current_ways, current_score)
                best_ways = ways  # Update total ways (accumulate)
                best_score = max(best_score, score)  # Keep max score
                existing_hats.remove(hat)  # Fixed typo: existin_hats -> existing_hats

        return [best_ways, best_score]

    result = [0, 0]
    result = dfs(0, 0, 0)  # Start with 0 ways and 0 score
    return result
```

```python
class Solution:
    def numberWays(self, hats: List[List[int]]) -> int:
        MOD = 10**9 + 7
        n = len(hats)
        
        # Create a mapping: hat -> list of people who can wear it
        hat_to_people = [[] for _ in range(41)]
        for person in range(n):
            for hat in hats[person]:
                hat_to_people[hat].append(person)
        
        # dp[mask] = number of ways to assign hats to people in the mask
        dp = [0] * (1 << n)
        dp[0] = 1  # Base case: 1 way to assign 0 hats to 0 people
        
        # Iterate through each hat
        for hat in range(1, 41):
            # If no one can wear this hat, skip it
            if not hat_to_people[hat]:
                continue
            
            # Process states in reverse order to avoid using updated values
            for mask in range((1 << n) - 1, -1, -1):
                if dp[mask] == 0:
                    continue
                    
                # Try assigning this hat to each person who likes it
                for person in hat_to_people[hat]:
                    # Check if this person doesn't have a hat yet
                    if mask & (1 << person) == 0:
                        new_mask = mask | (1 << person)
                        dp[new_mask] = (dp[new_mask] + dp[mask]) % MOD
        
        # Return the number of ways where all n people have hats
        return dp[(1 << n) - 1]
```

```python
def calculate(total_players, total_hats, owned_hat_length, hat_list, hat_points):
    if total_players > total_hats:
        return [0, 0]

    existing_hats = set()  # Fixed typo: exisiting -> existing
    total_ways = [0]  # Use list to modify in nested function
    max_score = [0]   # Track maximum score globally
    
    def dfs(player):  # Removed current_ways and current_score parameters
        if player == total_players:
            # When all players have hats, count this way and calculate score
            total_ways[0] += 1
            # Calculate score for this assignment
            score = 1
            for wore_hat in existing_hats:
                score *= hat_points[wore_hat]
            max_score[0] = max(max_score[0], score)
            return

        for hat in hat_list[player]:
            if hat not in existing_hats:
                existing_hats.add(hat)
                dfs(player + 1)  # Continue to next player
                existing_hats.remove(hat)  # Fixed typo: existin_hats -> existing_hats

    dfs(0)  # Start from player 0
    return [total_ways[0], max_score[0]]
```
