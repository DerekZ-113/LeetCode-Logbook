## ðŸ” Median of Two Sorted Arrays â€“ Description & Approach

**Problem**  
Given two individually **sorted** arrays `nums1` and `nums2` (lengths `m` and `n`), return the **median** of their union in **O(log min(m, n))** time.

### âš™ï¸ Core Ideaâ€‚â€”â€‚Partition, Donâ€™t Merge
1. **Always binary-search the shorter array** (`nums1`) to minimize the search window.  
2. Choose a cut index `i` in `nums1`; derive the matching cut `j` in `nums2` so the **left half** contains exactly  
   \[(m + n + 1) / 2\] elements (the left side takes the extra one when the total is odd).  
   `j = (m + n + 1) // 2 âˆ’ i`  
3. Define border values with sentinels for edge cuts:  
   * `L1 = nums1[i-1]` or âˆ’âˆžâ€ƒ`R1 = nums1[i]` or +âˆž  
   * `L2 = nums2[j-1]` or âˆ’âˆžâ€ƒ`R2 = nums2[j]` or +âˆž  
4. **Valid-partition test**  
   `L1 â‰¤ R2 and L2 â‰¤ R1`  
   If both hold, every element on the left is â‰¤ every element on the right.  
5. **Median extraction**  
   * Odd totalâ€‚â†’ `max(L1, L2)`  
   * Even totalâ€‚â†’ `(max(L1, L2) + min(R1, R2)) / 2`  
6. If the test fails, shift the search window:  
   * `L1 > R2` â†’ cut too far right â†’ `high = i âˆ’ 1`  
   * otherwise (`L2 > R1`) â†’ cut too far left â†’ `low = i + 1`

### ðŸ“ˆ Complexity
* **Time:** O(log min(m, n))  
* **Space:** O(1)

### âœ… Why It Works
Because each input array is sorted, checking only the four border values (`L1`, `R1`, `L2`, `R2`) is **both necessary and sufficient** to guarantee global order across the partition. Once `L1 â‰¤ R2` and `L2 â‰¤ R1`, the borders lock the two halves into sorted order, letting us read the median directly from those boundary elements.

## ðŸ§ª Example Walkthrough: Median of Two Sorted Arrays

Given:

- `nums1 = [1, 3]`
- `nums2 = [2, 5, 6, 7]`

We apply binary search on the **shorter array** `nums1`.

- Total elements = 6 (even)
- Goal: split arrays so that elements on the left and right form two equal halves  
- Then take the average of the max of left and min of right

---

### ðŸ” Iteration 1

- `left = 0`, `right = 2`
- `cut1 = (0 + 2) // 2 = 1`
- `cut2 = ((2 + 4 + 1) // 2) - 1 = 2`

**Partitions:**

```
nums1: [1] | [3]
nums2: [2, 5] | [6, 7]
```

**Boundary values:**

- `L1 = nums1[cut1 - 1] = 1`
- `R1 = nums1[cut1]     = 3`
- `L2 = nums2[cut2 - 1] = 5`
- `R2 = nums2[cut2]     = 6`

**Checks:**

- âœ… `L1 â‰¤ R2`: `1 â‰¤ 6`
- âŒ `L2 â‰¤ R1`: `5 â‰¤ 3` â†’ **Invalid**

**Action:**
Move `cut1` right â†’ `left = 2`

---

### ðŸ” Iteration 2

- `left = 2`, `right = 2`
- `cut1 = (2 + 2) // 2 = 2`
- `cut2 = 3 - 2 = 1`

**Partitions:**

```
nums1: [1, 3] | []
nums2: [2] | [5, 6, 7]
```

**Boundary values:**

- `L1 = nums1[cut1 - 1] = 3`
- `R1 = âˆž` (nothing right of cut in nums1)
- `L2 = nums2[cut2 - 1] = 2`
- `R2 = nums2[cut2]     = 5`

**Checks:**

- âœ… `L1 â‰¤ R2`: `3 â‰¤ 5`
- âœ… `L2 â‰¤ R1`: `2 â‰¤ âˆž` â†’ **Valid partition found**

### ðŸ§® Final Median Calculation

Since total length is **even**, take the average of:

- `max(L1, L2) = max(3, 2) = 3`
- `min(R1, R2) = min(âˆž, 5) = 5`

**Median = (3 + 5) / 2 = 4.0**

### âœ… Final Answer: `4.0`

````python
class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        # Ensure nums1 is the smaller array for minimal binary search range
        if len(nums1) > len(nums2):
            nums1, nums2 = nums2, nums1

        n1, n2 = len(nums1), len(nums2)

        # Binary search boundaries on the smaller array
        left, right = 0, n1

        while left <= right:
            # Partition index for nums1
            cut1 = (left + right) // 2
            # Partition index for nums2 so that left partitions add up to half
            cut2 = ((n1 + n2 + 1) // 2) - cut1

            # Handle edge cases where the partition cuts at the ends
            L1 = float("-inf") if cut1 == 0 else nums1[cut1 - 1]  # Left side max of nums1
            R1 = float("inf") if cut1 == n1 else nums1[cut1]      # Right side min of nums1

            L2 = float("-inf") if cut2 == 0 else nums2[cut2 - 1]  # Left side max of nums2
            R2 = float("inf") if cut2 == n2 else nums2[cut2]      # Right side min of nums2

            # Check if correct partition is found
            if L1 <= R2 and L2 <= R1:
                # Odd total length: median is the max of left parts
                if (n1 + n2) % 2 == 1:
                    return float(max(L1, L2))
                # Even total length: median is the average of mid two values
                else:
                    return (max(L1, L2) + min(R1, R2)) / 2.0

            # Too far right in nums1, move left
            if L1 > R2:
                right = cut1 - 1
            # Too far left in nums1, move right
            else:
                left = cut1 + 1
````
