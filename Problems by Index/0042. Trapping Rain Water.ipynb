## üìê Description / Approach

**Goal**  
Given an elevation map `height`, compute the total water trapped after rain.

**Core insight**  
At any index *i*, the water level is limited by the shorter of  
* `left_max` ‚Äî highest bar seen so far from the **left**, and  
* `right_max` ‚Äî highest bar seen so far from the **right**.

**Two-pointer O(n) / O(1) algorithm**  
1. **Pointers** ‚Äì start `left = 0`, `right = n-1`.  
2. **Running maxima** ‚Äì track `left_max`, `right_max` as the tallest walls found so far on each side.  
3. **Iterate while `left < right`:**  
   * If `left_max ‚â§ right_max`  
     * Move `left` one step inwards.  
     * Update `left_max = max(left_max, height[left])`.  
     * Water at this index = `left_max ‚Äì height[left]` (never negative).  
   * Else (right side is lower or equal)  
     * Move `right` one step inwards.  
     * Update `right_max = max(right_max, height[right])`.  
     * Water at this index = `right_max ‚Äì height[right]`.  
4. Accumulate all water values in `result`.

**Why update the max *before* adding water?**  
A bar that resets the record height is a **wall**, not a **bucket**.  
Updating first eliminates negative additions and naturally yields *0* water on peaks.

Time = O(n) (single pass)‚ÄÉSpace = O(1) (four scalars). üöÄ

```python
from typing import List

class Solution:
    def trap(self, height: List[int]) -> int:
        """
        Compute how much water can be trapped after raining.

        Two-pointer / O(1)-space approach:
        - Keep a pointer at each end (`left`, `right`).
        - Track the tallest bar seen so far from each side
            (`left_max`, `right_max`).
        - Always move the pointer on the **shorter** side, because that
            side‚Äôs max height is the current limiting wall.
        - Water above the moved pointer = (that side‚Äôs max) ‚àí (bar height).
        """
        n = len(height)
        if n < 3:                # ‚â§2 bars can never trap water
            return 0

        # --- Initialization -----------------------------
        left, right = 0, n - 1               # two pointers
        left_max, right_max = height[left], height[right]
        water = 0                            # running total

        # --- Main sweep ---------------------------------
        while left < right:
            if left_max <= right_max:
                # Resolve the *left* side because its max wall is shorter
                left += 1
                left_max = max(left_max, height[left])        # update record
                water += left_max - height[left]              # add trapped water (‚â•0)
            else:
                # Resolve the *right* side
                right -= 1
                right_max = max(right_max, height[right])     # update record
                water += right_max - height[right]            # add trapped water (‚â•0)

        return water
```

## üß† Key Concepts Recap

- **Water formula**: `water[i] = min(left_max, right_max) ‚Äì height[i]`.  
- Always move the pointer at the **shorter side**; the taller side guarantees a sealing wall.  
- Update `left_max` / `right_max` **before** computing trapped water to avoid negatives.  
- Edge cases: empty / single bar ‚Üí 0; strictly increasing or decreasing arrays ‚Üí 0.  
- Alternatives worth knowing:  
  - Prefix‚Äìsuffix max arrays (O(n) space, simpler reasoning).  
  - Monotonic stack (O(n) space, ‚Äúfill valleys‚Äù method).  
- Interview flex üí™: two-pointer version shows mastery of constant-space optimization.
