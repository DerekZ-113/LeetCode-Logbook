## 1296. Divide Array in Sets of K Consecutive Numbers

### 📝 Description
Given an array of integers `nums` and an integer `k`, return `True` if it is possible to divide this array into sets of `k` **consecutive** numbers. Otherwise, return `False`.

Each element in `nums` must be used exactly once.

---

### ⚙️ Approach
Use a **greedy + heap + frequency counter** approach to always form the lowest-possible valid group:

1. **Quick check**: If `len(nums) % k != 0`, return `False` immediately.
2. Use a `Counter` to count the frequency of each number.
3. Create a **min-heap** of the unique keys in `nums` to always start forming sequences from the smallest available number.
4. While the heap is not empty:
   - Take the smallest number `first`.
   - Try to form a sequence `[first, first+1, ..., first+k-1]`.
   - For each number `i` in this range:
     - If `i` is not in `count`, return `False`.
     - Decrease the count.
     - If `count[i] == 0`:
       - It **must** be at the top of the heap to maintain heap integrity — otherwise return `False`.
       - Pop it from the heap.

---

### 🧠 Key Concepts
- **Min Heap**:
  - Ensures you always start forming sequences from the smallest available number.
- **Greedy Grouping**:
  - Form groups immediately when the lowest number is available.
- **Heap Validity Check**:
  - Maintain consistency by ensuring only the heap's root is removed when depleted.
- **Time Complexity**: O(n log n) due to heap operations.
- **Space Complexity**: O(n) for the counter and heap.

---

### 🔍 Example
```python
Input: nums = [1,2,3,3,4,4,5,6], k = 4

Output: True
Explanation: The array can be divided into:
- [1,2,3,4]
- [3,4,5,6]

````python
class Solution:
    def isPossibleDivide(self, nums: List[int], k: int) -> bool:
        if len(nums) % k:
            return False

        count = Counter(nums)
        heap = list(count.keys())
        heapq.heapify(heap)

        while heap:
            first = heap[0]
            for i in range(first, first + k):
                if i not in count:
                    return False
                count[i] -= 1
                if count[i] == 0:
                    if i != heap[0]:
                        return False
                    heapq.heappop(heap)

        return True
````
