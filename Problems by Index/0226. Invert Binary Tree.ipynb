## 226. Invert Binary Tree

### ðŸ“ Description
Given the `root` of a binary tree, invert the tree, and return its root.

**Inversion** means flipping the tree around its center â€” swapping every left and right child in the tree.

---

### âš™ï¸ Approach
Use **recursion** to invert the tree in a top-down manner:

1. **Base Case**:  
   - If the node is `None`, return immediately.
2. **Swap** the left and right children.
3. **Recurse** on the (now swapped) left and right children.
4. Return the modified root.

This approach inverts the entire tree in-place.

---

### ðŸ§  Key Concepts
- **Recursive Tree Traversal**:
  - This is a pre-order traversal with in-place modification.
- **In-place Mutation**:
  - No need to create a new tree â€” we reuse existing nodes.
- **Time Complexity**: O(n)  
  - Every node is visited once.
- **Space Complexity**: O(h)  
  - `h` is the height of the tree (due to recursion stack).

---

### ðŸ” Example
```python
Input:
    4
   / \
  2   7
 / \ / \
1  3 6  9

Output:
    4
   / \
  7   2
 / \ / \
9  6 3  1

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        
        def invert(node):
            if not node:
                return
            node.left, node.right = node.right, node.left
            invert(node.left)
            invert(node.right)

        invert(root)
        return root
````
